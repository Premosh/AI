```python
# Define the environment (Rooms and their states)
class Environment:
   def __init__(self):
       # Assume the environment consists of two rooms: A and B
       self.location_condition = {'A': 'Dirty', 'B': 'Clean'}


# Define the Agent (Vacuum Cleaner)
class SimpleReflexAgent:
   def __init__(self, environment):
       # Initial location of the vacuum cleaner
       self.location = 'A'
       self.environment = environment

   # Define the perception of the agent (sense environment)
   def perceive(self):
       return self.environment.location_condition[self.location]


   # Define the action based on the perception
   def act(self):
       if self.perceive() == 'Dirty':
           print(f"Location {self.location} is Dirty. Cleaning...")
           self.environment.location_condition[self.location] = 'Clean'
           print(f"Location {self.location} is now Clean.")
       else:
           print(f"Location {self.location} is Clean. Moving to next location.")
           # Move to the other room
           self.location = 'B' if self.location == 'A' else 'A'


# Simulation
def run_agent():
   # Initialize environment and agent
   env = Environment()
   agent = SimpleReflexAgent(env)

   # Run the agent's decision-making for 5 steps
   for step in range(5):
       print(f"\nStep {step+1}:")
       agent.act()

run_agent()
```
